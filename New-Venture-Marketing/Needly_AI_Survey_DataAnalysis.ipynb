#Data Analysis

```python
# Analyze Shopper Frequency

# Grouping by shopping frequency to analyze key metrics
shopping_frequency_analysis = survey_df.groupby(
    'Shopper Frequency'
).agg({
    'Likelihood to Use': ['mean', 'count'],
    'Likelihood to Recommend': 'mean',
    'AI Usefulness': lambda x: x.value_counts(normalize=True).to_dict(),
    'Appealing Features': lambda x: x.value_counts(normalize=True).to_dict(),
    'Hesitations': lambda x: x.value_counts(normalize=True).to_dict()
})

shopping_frequency_analysis.columns = [
    'Average Likelihood to Use', 'Response Count', 'Average Likelihood to Recommend',
    'AI Usefulness Distribution', 'Appealing Features Distribution', 'Hesitation Factors Distribution'
]

# Display the results
print(shopping_frequency_analysis)
```

#Visualization

```python
# Data for visualization
shopping_frequency = shopping_frequency_analysis.index
avg_likelihood_use = shopping_frequency_analysis['Average Likelihood to Use']
avg_likelihood_recommend = shopping_frequency_analysis['Average Likelihood to Recommend']

# Create a bar chart for average likelihood to use and recommend
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width
bar_width = 0.35
x = range(len(shopping_frequency))

# Plot bars
ax.bar(x, avg_likelihood_use, bar_width, label='Likelihood to Use', alpha=0.8)
ax.bar([i + bar_width for i in x], avg_likelihood_recommend, bar_width, label='Likelihood to Recommend', alpha=0.8)

# Add labels and title
ax.set_xlabel('Shopper Frequency', fontsize=12)
ax.set_ylabel('Average Score', fontsize=12)
ax.set_title('Likelihood to Use and Recommend Needly by Shopping Frequency', fontsize=14)
ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(shopping_frequency, rotation=45, ha='right')
ax.legend()

# Show the chart
plt.tight_layout()
plt.show()
```

```python
# Grouping data by gender and AI usefulness
ai_usefulness_gender = survey_df.groupby('Gender')['AI Usefulness'].value_counts(normalize=True).unstack()

# Plotting the data
fig, ax = plt.subplots(figsize=(8, 6))

ai_usefulness_gender.plot(kind='bar', stacked=True, ax=ax, alpha=0.8)

# Adding labels and title
ax.set_title('AI Usefulness by Gender', fontsize=14)
ax.set_xlabel('Gender', fontsize=12)
ax.set_ylabel('Proportion', fontsize=12)
ax.legend(title='AI Usefulness', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_xticklabels(ai_usefulness_gender.index, rotation=0)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

The chart above visualizes the perceived usefulness of AI-driven predictions for restocking needs by gender. It shows the proportions of responses (e.g., "Yes," "No," "Maybe") for each gender group.

```python
# Grouping data by occupation and AI usefulness
ai_usefulness_occupation = survey_df.groupby('Occupation')['AI Usefulness'].value_counts(normalize=True).unstack()

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 6))

ai_usefulness_occupation.plot(kind='bar', stacked=True, ax=ax, alpha=0.8)

# Adding labels and title
ax.set_title('AI Usefulness by Occupation', fontsize=14)
ax.set_xlabel('Occupation', fontsize=12)
ax.set_ylabel('Proportion', fontsize=12)
ax.legend(title='AI Usefulness', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_xticklabels(ai_usefulness_occupation.index, rotation=45, ha='right')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

The chart above illustrates the perceived usefulness of AI-driven predictions for restocking needs by occupation. Each bar represents the proportion of responses ("Yes," "No," "Maybe") within each occupation group.

```python
# Grouping data by gender and preferred features
preferred_features_gender = survey_df.groupby('Gender')[
    'Appealing Features'
].value_counts(normalize=True).unstack()

# Plotting the data
fig, ax = plt.subplots(figsize=(12, 6))

preferred_features_gender.plot(kind='bar', stacked=True, ax=ax, alpha=0.8)

# Adding labels and title
ax.set_title('Preferred Features by Gender', fontsize=14)
ax.set_xlabel('Gender', fontsize=12)
ax.set_ylabel('Proportion', fontsize=12)
ax.legend(title='Preferred Features', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_xticklabels(preferred_features_gender.index, rotation=0)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

The chart above displays the distribution of preferred features in an AI-driven subscription model by gender. Each bar represents the proportion of responses for different features within each gender group.

```python
# Calculate the overall distribution of hesitation factors
hesitation_factors = survey_df['Hesitations'].value_counts(normalize=True)

# Plotting the data
fig, ax = plt.subplots(figsize=(8, 6))

hesitation_factors.plot(kind='bar', ax=ax, alpha=0.8)

# Adding labels and title
ax.set_title('Most Common Hesitations Overall', fontsize=14)
ax.set_xlabel('Hesitation Factors', fontsize=12)
ax.set_ylabel('Proportion', fontsize=12)
ax.set_xticklabels(hesitation_factors.index, rotation=45, ha='right')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

The chart above highlights the most common hesitation factors overall for using a service like Needly. It shows the proportion of respondents who cited each hesitation.

```python
# Grouping data by occupation and hesitation factors
hesitations_by_occupation = survey_df.groupby('Occupation')[
    'Hesitations'
].value_counts(normalize=True).unstack()

# Plotting the data
fig, ax = plt.subplots(figsize=(12, 6))

hesitations_by_occupation.plot(kind='bar', stacked=True, ax=ax, alpha=0.8)

# Adding labels and title
ax.set_title('Common Hesitations by Occupation', fontsize=14)
ax.set_xlabel('Occupation', fontsize=12)
ax.set_ylabel('Proportion', fontsize=12)
ax.legend(title='Hesitation Factors', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_xticklabels(hesitations_by_occupation.index, rotation=45, ha='right')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

The chart above illustrates the common hesitation factors for using Needly, broken down by occupation. Each bar represents the proportion of respondents within each occupation group who selected a specific hesitation.

```python
# Grouping data by occupation and preferred features
preferred_features_by_occupation = survey_df.groupby('Occupation')[
    'Appealing Features'
].value_counts(normalize=True).unstack()

# Plotting the data
fig, ax = plt.subplots(figsize=(12, 6))

preferred_features_by_occupation.plot(kind='bar', stacked=True, ax=ax, alpha=0.8)

# Adding labels and title
ax.set_title('Preferred Features by Occupation', fontsize=14)
ax.set_xlabel('Occupation', fontsize=12)
ax.set_ylabel('Proportion', fontsize=12)
ax.legend(title='Preferred Features', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_xticklabels(preferred_features_by_occupation.index, rotation=45, ha='right')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

The chart above compares the preferred features of an AI-driven subscription model across different occupations.

Each bar represents the proportion of respondents within each occupation group who selected a specific feature as appealing.

**AI Usefulness:**

**Shopping Behavior:**
